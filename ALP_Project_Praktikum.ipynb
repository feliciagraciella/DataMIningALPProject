## Data Preprocessing

### Missing Values

Menampilkan jumlah missing values pada masing-masing kolom

```python
df.isnull().sum()
```

Menampilkan data dengan missing values

```python
df[df.isna().any(axis=1)]
```

Ada beberapa cara mengisi data null yaitu 
1. Menghapus Row
2. Menghapus kolom
3. Mean : digunakan ketika datanya terdistribusi normal
4. Median : digunakan ketika datanya skewed.
5. Mode : digunakan untuk data kategorikal.
6. Menggunakan Regresi ketika datanya kontinu
7. Menggunakan Classifier ketika datanya kategorikal

Handling Missing Value
1. time menggunakan median karena datanya cenderung ke skewed
2. warts_count menggunakan median karena datanya skewed
3. type menggunakan mode karena kategorikal
4. induration diameter menggunakan median kerena datanya skewed

#### Time

```python
df['time'].fillna(df['time'].median(), inplace = True)
```

#### Number of Warts

```python
df['warts_count'].fillna(df['warts_count'].median(), inplace = True)
```

#### Induration Diameter

```python
df['induration_diameter'].fillna(df['induration_diameter'].median(), inplace = True)
```

#### Type

```python
typemode = df['type'].mode()
df['type'].fillna(typemode[0], inplace = True)
```

## Classification

### Imbalanced Data

```python
X = df_final.iloc[:, :-1].values
y = df_final.iloc[:, -1].values
```

```python
from imblearn.over_sampling import SMOTE

over_sample = SMOTE()
X, y = over_sample.fit_resample(X, y)
```

```python
from collections import Counter

counter = Counter(y)
for k,v in counter.items():
	per = v / len(y) * 100
	print('Class=%d, n=%d (%.3f%%)' % (k, v, per))

plt.bar(counter.keys(), counter.values())
plt.title('Oversampling with SMOTE', pad=20)
plt.xlabel('Result of Treatment')
plt.ylabel('Number of Sample after SMOTE')
plt.show()
```

```python
Xclass = df_final.iloc[:,:-1].values
yclass = df_final.iloc[:, -1].values
```

```python
Xclass_train, Xclass_test, yclass_train, yclass_test = train_test_split(Xclass, yclass, test_size = 0.1)
```

```python
nb_yclass_pred = nbclassifier.predict(Xclass_test)
print(nb_yclass_pred)
```

```python
nbcm = confusion_matrix(yclass_test, nb_yclass_pred)
print("Confusion Matrix\n", nbcm)

print("Accuracy Score: ", accuracy_score(yclass_test, nb_yclass_pred))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rforest = RandomForestClassifier(random_state=False, n_jobs=-1)
rforest.fit(Xclass_train, yclass_train)
```

```python
rf_yclass_pred = rforest.predict(Xclass_test)
print(rf_yclass_pred)
```

## Clustering

### Scaling

```python
sc = StandardScaler()
scaled = sc.fit_transform(dfclustering)
scaleddf = pd.DataFrame(scaled)
scaleddf.head()
```

### K means

```python
from sklearn.cluster import KMeans

distortions = []
K = range(1,11)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(scaleddf)
    distortions.append(kmeanModel.inertia_)
    print("K =", k, " ", distortions[k-1])
```

```python
# manually visualize the elbow method using plotlib
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.xticks(np.arange(1,11,1))
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
from kneed import DataGenerator, KneeLocator

knee = KneeLocator(K, distortions, curve = "convex", direction = "decreasing")
print("The optimal k for the model is", knee.elbow)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 4, random_state = 1)
y_kmeans = kmeans.fit_predict(scaleddf)
y_kmeans
```

```python
dfclustering['K means'] = y_kmeans
dfclustering.groupby('K means').agg(Gender = ('gender', pd.Series.mode), 
                          Age = ('age', 'mean'), 
                          Time = ('time', 'mean'), 
                          Warts_Count = ('warts_count', 'mean'), 
                          Type = ('type', pd.Series.mode), 
                          Area = ('area', 'mean'), 
                          Induration_Diameter = ('induration_diameter', 'mean'), 
                          Treatment_Result = ('treatment_result', pd.Series.mode), 
                          Count = ('K means', 'count'))
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(scaleddf,y_kmeans)
```

### Hierarchical

```python
import scipy.cluster.hierarchy as sch

plt.figure(figsize=(16,5))
dist_matrix = sch.linkage(scaleddf, method='complete')
dn = sch.dendrogram(dist_matrix)
plt.title('Dendogram')
plt.show
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean')
y_hc = hc.fit_predict(scaleddf)
y_hc
```

```python
dfclustering['Hierarchical Cluster'] = y_hc
dfclustering.groupby('Hierarchical Cluster').agg(Gender = ('gender', pd.Series.mode), 
                          Age = ('age', 'mean'), 
                          Time = ('time', 'mean'), 
                          Warts_Count = ('warts_count', 'mean'), 
                          Type = ('type', pd.Series.mode), 
                          Area = ('area', 'mean'), 
                          Induration_Diameter = ('induration_diameter', 'mean'), 
                          Treatment_Result = ('treatment_result', pd.Series.mode), 
                          Count = ('Hierarchical Cluster', 'count'))
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(scaleddf,y_hc)
```
